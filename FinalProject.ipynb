### 2.1. The Persistent Nature of Gender Bias in Media

The issue of gender bias in media reporting has been consistently documented across different political contexts. For example, Atkeson and Krebs (2008) examined gender bias in media coverage of mayoral candidates. They found that while press coverage was not overtly biased in favor of male candidates, there were subtle differences in how male and female candidates were framed. Their study emphasized the role of gendered issue framing, with male candidates often receiving more positive coverage regarding leadership qualities and policy expertise.

In a meta-analysis, VanderPas and Aaldering (2023) analyzed over 90 studies covering 25,000 politicians and found that women in proportional electoral systems consistently received less media coverage than their male counterparts. Interestingly, this gender gap was absent in majoritarian electoral systems. The study also revealed that women were more likely to be covered in relation to personal life and family matters, while men were more often associated with political competence and leadership.

Leavy (2005) utilized machine learning and natural language processing to uncover gender biases in the media coverage of Irish political leaders over a 15-year period. The findings suggested that female politicians were more often associated with social policies and portrayed less favorably in terms of their political performance. Leavy's study highlighted the potential for using AI tools to identify hidden gender biases in large-scale media content analysis.

Bligh (2011) examined how gender stereotypes influenced the media’s portrayal of female politicians. Her study found that female politicians were often depicted as competent but "cold" or "unlikable," with an emphasis on their personality traits rather than their political qualifications. This framing of female politicians contributes to a perception that, while they may be capable leaders, they are not as approachable or relatable as their male counterparts.

These studies collectively highlight the ongoing nature of gender bias in political media coverage. Despite progress in gender equality in politics, the media continues to play a significant role in reinforcing stereotypes and shaping public perceptions of female politicians.

```python
def parse_factiva(file_path):
    """
    Parses Factiva HTML to extract articles with headlines marked by <b>HD</b>,
    authors marked by <b>BY</b>, and content within <p class="articleParagraph">.
    """
    articles = []
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    for hd_tag in soup.find_all('b', string='HD'):
        # Extract headline from the next sibling's span with class 'enHeadline'
        headline_span = hd_tag.find_next('span', class_='enHeadline')
        headline = headline_span.get_text(strip=True) if headline_span else None

        # Extract author from the 'BY' marker
        by_tag = hd_tag.find_next('b', string='BY')
        author = by_tag.find_next('td').get_text(strip=True) if by_tag else None

        # Extract article content in <p class="articleParagraph"> until the next headline
        content = []
        current_tag = hd_tag.find_parent('tr').find_next_sibling('tr')
        while current_tag:
            # Stop if another HD is encountered
            next_hd = current_tag.find('b', string='HD')
            if next_hd:
                break
            # Extract <p> with article content
            paragraphs = current_tag.find_all('p', class_='articleParagraph')
            content.extend([p.get_text(strip=True) for p in paragraphs])
            current_tag = current_tag.find_next_sibling('tr')

        # Append extracted data
        if headline or content:
            articles.append({
                'headline': headline,
                'author': author,
                'content': ' '.join(content)
            })

    return articles
```

```python
def process_factiva_files(file_paths, names):
    results = {}

    for file_path, name in zip(file_paths, names):
        # Parse the file
        articles = parse_factiva(file_path)

        # Convert to DataFrame
        df = pd.DataFrame(articles)

        # Store in results dictionary
        results[name] = df
        print(f"Processed {name} with {len(df)} articles.")

    return results
```

Now let's utilize regular expression, NLTK's name entity recognition as well as stop words to filter out unimportant details that will skew our analysis. Here we will explicitly filter out generic political adjectives like 'federal', 'prime', 'general'; as well as gender specific pronouns that addressed the politician themselves.

```python
filter_words = {'federal', 'prime', 'general', 'political',
                'said', 'like','says','told','say','go',
                'year','updatedtimeupdated','new', 'bst',
                'publishedtime','blocktime','gmt','updatedtime', 'updated', 'pm', 'am',
                'mr', 'ms', 'she', 'her', 'he', 'him',}
custom_stopwords = set(stopwords.words('english')).union(filter_words)
```

The clean_text function will help us filter for only important adjectives as well as relevant entities.

```python
# Function to clean text and extract relevant entities
def clean_text(text):
    """
    Cleans text by removing URLs, special characters, and stopwords, and performing Named Entity Recognition.
    Args:
        text (str): Input text to clean.
    Returns:
        str: Cleaned text with relevant entities and words.
    """
    if not isinstance(text, str):
        return ""

    # Remove URLs and non-alphabetic characters
    text = re.sub(r'http\S+|www\S+|\.com', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize and remove stopwords
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in custom_stopwords and len(word) > 1]

    # Named Entity Recognition (NER) to keep relevant entities
    tagged_tokens = pos_tag(filtered_tokens)
    named_entities = ne_chunk(tagged_tokens)
    relevant_words = []

    for chunk in named_entities:
        if hasattr(chunk, 'label') and chunk.label() in ['ORG', 'EVENT', 'LAW']:
            relevant_words.extend([token[0] for token in chunk.leaves()])
        elif isinstance(chunk, tuple):
            relevant_words.append(chunk[0])

    return ' '.join(relevant_words)
```

```python
# Function to extract adjectives from text
def extract_adjectives(text):
    """
    Cleans text by removing URLs, special characters, stopwords, and extracts adjectives only.
    Args:
        text (str): Input text to process.
    Returns:
        str: Text containing only adjectives.
    """
    if not isinstance(text, str):
        return ""

    # Remove URLs and non-alphabetic characters
    text = re.sub(r'http\S+|www\S+|\.com', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize and remove stopwords
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in custom_stopwords and len(word) > 1]

    # Extract adjectives using POS tagging
    adjectives = [word for word, pos in pos_tag(filtered_tokens) if pos in ['JJ', 'JJR', 'JJS']]
    return ' '.join(adjectives)
```

```python
# Apply cleaning and adjective extraction to each politician's DataFrame
for name, df in factiva_dataframes.items():
    print(f"\nProcessing text data for {name}")
    # Apply clean_text to extract entities
    df['content_entities'] = df['content'].apply(clean_text)
    # Apply extract_adjectives to extract adjectives
    df['content_adjectives'] = df['content'].apply(extract_adjectives)
```

### 4.1 TF-IDF (Term Frequency-Inverse Document Frequency):

TFIDF (Term Frequency-Inverse Document Frequency) is a technique that will help identify unique words and phrases associated with each politician, revealing patterns of gendered framing in the media.

$$\text{TF-IDF} = \text{term_frequency} * \text{inverse_document_frequency}$$

  Where:

  $
  \text{term_frequency} = \text{number of times a given term appears in document}
  \\
  \text{inverse_document_frequency} = log(\text{total number of documents} / \text{number of documents with term}) + 1
  $

#### 4.1.1 TF-IDF Analysis for Entities

First, we define a set of potentially gender-specific words.

```python
# Separate and analyze entities for male and female politicians
male_terms = []
female_terms = []

male_names = {'donald', 'trump', 'rishi', 'sunak'}
female_names = {'kamala', 'harris', 'elizabeth', 'truss'}

for name, df in factiva_dataframes.items():
    print(f"\nTF-IDF Analysis of entities- {name}")

    # Combine all adjectives into a single document
    combined_text = ' '.join(df['content_entities'].dropna())

    if not combined_text.strip():
        print(f"No content available for {name}. Skipping...")
        continue

    # Adjust TF-IDF parameters for clean text
    vectorizer = TfidfVectorizer(stop_words='english', max_df=1.0, min_df=1)

    try:
        # Transform the text using TF-IDF
        tfidf_matrix = vectorizer.fit_transform([combined_text])
        feature_names = vectorizer.get_feature_names_out()

        # Extract TF-IDF scores and sort them
        scores = np.array(tfidf_matrix.sum(axis=0)).flatten()
        tfidf_scores = dict(zip(feature_names, scores))
        sorted_scores = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:50]

        # Identify gendered terms
        gender_bias_terms = {term: score for term, score in sorted_scores if term in gendered_terms}

        print("Top 10 TF-IDF terms (Entities):")
        for term, score in sorted_scores[:10]:
            print(f"{term}: {score:.4f}")

        if gender_bias_terms:
            print("\nIdentified Gendered Terms:")
            for term, score in gender_bias_terms.items():
                print(f"{term}: {score:.4f}")

        # Separate terms by gender
        if any(name.lower() in female_names for name in name.split()):
            female_terms.extend([term for term, score in sorted_scores])
        elif any(name.lower() in male_names for name in name.split()):
            male_terms.extend([term for term, score in sorted_scores])

        # Generate and display a word cloud
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(sorted_scores))
        plt.figure(figsize=(10, 6))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.title(f"Word Cloud for entities - {name}")
        plt.show()
    except ValueError as e:
        print(f"TF-IDF failed for {name}: {e}")
```

```python
def calculate_tfidf_scores(df, column, n_top=20):
    """
    Calculate TF-IDF scores for a specific column in the DataFrame.

    Args:
        df (pd.DataFrame): DataFrame containing text data.
        column (str): Column to analyze with TF-IDF.
        n_top (int): Number of top terms to return.

    Returns:
        dict: Sorted TF-IDF terms with their scores.
    """
    vectorizer = TfidfVectorizer(stop_words='english', max_df=0.85, min_df=1)
    documents = df[column].dropna().tolist()
    if not documents:
        return {}

    tfidf_matrix = vectorizer.fit_transform(documents)
    scores = np.mean(tfidf_matrix.toarray(), axis=0)
    terms = vectorizer.get_feature_names_out()
    tfidf_scores = dict(zip(terms, scores))
    return dict(sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:n_top])
```

```python
def analyze_and_plot_tf_idf(tfidf_male, tfidf_female, gender_specific_terms):
    """
    Analyzes TF-IDF scores for male and female politicians, highlighting gender-specific terms.
    Args:
        tfidf_male (dict): TF-IDF scores for male politicians.
        tfidf_female (dict): TF-IDF scores for female politicians.
        gender_specific_terms (set): Predefined gender-specific topics or words.
    """
    # Convert TF-IDF scores into DataFrames
    male_df = pd.DataFrame(tfidf_male.items(), columns=['Term', 'Score']).sort_values(by='Score', ascending=False).head(20)
    female_df = pd.DataFrame(tfidf_female.items(), columns=['Term', 'Score']).sort_values(by='Score', ascending=False).head(20)

    # 1. Plot Top TF-IDF Words
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    sns.barplot(data=male_df, x='Score', y='Term', ax=axes[0], color='#F28E2B')
    axes[0].set_title("Top TF-IDF Words - Male Politicians")
    sns.barplot(data=female_df, x='Score', y='Term', ax=axes[1], color='skyblue')
    axes[1].set_title("Top TF-IDF Words - Female Politicians")
    plt.tight_layout()
    plt.show()

    # 2. Analyze Gender-Specific Terms (including zeros)
    male_specific_scores = {term: tfidf_male.get(term, 0) for term in gender_specific_terms}
    female_specific_scores = {term: tfidf_female.get(term, 0) for term in gender_specific_terms}

    # Combine results into a DataFrame
    combined_df = pd.DataFrame({
        'Term': list(male_specific_scores.keys()) + list(female_specific_scores.keys()),
        'Score': list(male_specific_scores.values()) + list(female_specific_scores.values()),
        'Gender': ['Male'] * len(male_specific_scores) + ['Female'] * len(female_specific_scores)
    })

    # Include zeros to emphasize absent mentions
    combined_df.sort_values(by='Score', ascending=False, inplace=True)

    # 3. Plot Gender-Specific Term Distribution (including zeros)
    plt.figure(figsize=(12, 8))
    sns.barplot(data=combined_df, x='Score', y='Term', hue='Gender', palette=['#F28E2B', 'skyblue'])
    plt.title("TF-IDF Scores of Gender-Specific Terms (Including Zero Mentions)")
    plt.xlabel("TF-IDF Score")
    plt.ylabel("Terms")
    plt.legend(title="Group")
    plt.show()
```

```python
def analyze_and_plot_tf_idf(male_terms, female_terms, gender_specific_terms):
    """
    Analyzes TF-IDF terms for male and female politicians and generates histograms, frequency plots, sentiment,
    and overlap visualizations.
    """
    # Count occurrences of terms
    male_term_counts = Counter(male_terms)
    female_term_counts = Counter(female_terms)

    # Create DataFrame for top terms
    male_df = pd.DataFrame(male_term_counts.items(), columns=['Term', 'Count']).sort_values(by='Count', ascending=False).head(20)
    female_df = pd.DataFrame(female_term_counts.items(), columns=['Term', 'Count']).sort_values(by='Count', ascending=False).head(20)

    # 1. Plot histograms of top terms
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    sns.barplot(data=male_df, x='Count', y='Term', ax=axes[0], color='#F3B562')  # Neutral blue
    axes[0].set_title("Top TF-IDF Words - Male Politicians")
    sns.barplot(data=female_df, x='Count', y='Term', ax=axes[1], color='skyblue')  # Light orange
    axes[1].set_title("Top TF-IDF Words - Female Politicians")
    plt.tight_layout()
    plt.show()

    # 2. Sentiment Analysis of Terms
    def calculate_sentiment(terms):
        sentiments = [TextBlob(word).sentiment.polarity for word in terms]
        return sum(sentiments) / len(sentiments) if sentiments else 0

    male_sentiment = calculate_sentiment(male_terms)
    female_sentiment = calculate_sentiment(female_terms)

    # Plot sentiment comparison
    plt.figure(figsize=(8, 5))
    sns.barplot(x=['Male Politicians', 'Female Politicians'], y=[male_sentiment, female_sentiment], palette=['#F3B562', 'lightblue'])
    plt.title("Average Sentiment of Terms for Male vs Female Politicians")
    plt.ylabel("Sentiment Polarity")
    plt.show()

    # 3. Frequency Plot for Gender-Specific Terms
    male_specific_counts = {term: male_term_counts[term] for term in gender_specific_terms if term in male_term_counts}
    female_specific_counts = {term: female_term_counts[term] for term in gender_specific_terms if term in female_term_counts}

    combined_df = pd.DataFrame({
        'Term': list(male_specific_counts.keys()) + list(female_specific_counts.keys()),
        'Count': list(male_specific_counts.values()) + list(female_specific_counts.values()),
        'Gender': ['Male'] * len(male_specific_counts) + ['Female'] * len(female_specific_counts)
    })

    plt.figure(figsize=(10, 6))
    sns.barplot(data=combined_df, x='Count', y='Term', hue='Gender', palette=['#F3B562', 'lightblue'])
    plt.title("Frequency of Gender-Specific Terms")
    plt.xlabel("Frequency")
    plt.ylabel("Terms")
    plt.legend(title="Group")
    plt.show()

    # 4. Overlap of Top Words
    male_top_terms = set(male_df['Term'])
    female_top_terms = set(female_df['Term'])

    plt.figure(figsize=(6, 6))
    venn = venn2([male_top_terms, female_top_terms], set_labels=('Male Politicians', 'Female Politicians'))
    plt.title("Overlap of Top TF-IDF Words")
    plt.show()
```

### 4.2. Topic Modeling:

Using Latent Dirichlet Allocation (LDA), this method will uncover the major themes in the articles, allowing the study to identify how female and male politicians are framed in different contexts.

By comparing articles about female and male politicians within each country and between the U.S. and U.K., the study will uncover regional differences in media coverage and reveal any gender biases in the portrayal of political figures.

```python
def perform_lda_topic_modeling(corpus, n_topics=3, n_top_words=10):
    """
    Performs Latent Dirichlet Allocation (LDA) on the given corpus to uncover major themes.
    Args:
        corpus (list): List of text documents.
        n_topics (int): Number of topics to extract.
        n_top_words (int): Number of top words to display per topic.
    """
    # Vectorize corpus
    vectorizer = CountVectorizer(stop_words='english', max_df=0.8, min_df=1)
    term_matrix = vectorizer.fit_transform(corpus)

    # Apply LDA
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
    lda.fit(term_matrix)

    # Display topics and top words
    terms = vectorizer.get_feature_names_out()
    print("\nTopics Identified:")
    for topic_idx, topic in enumerate(lda.components_):
        top_words = [terms[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        print(f"Topic {topic_idx + 1}: {', '.join(top_words)}")

    # Generate pyLDAvis visualization
    topic_term_dists = lda.components_ / lda.components_.sum(axis=1)[:, None]
    doc_topic_dists = lda.transform(term_matrix)
    doc_lengths = term_matrix.sum(axis=1).A1
    term_frequency = term_matrix.sum(axis=0).A1
    vocab = vectorizer.get_feature_names_out()
    vis = prepare(topic_term_dists, doc_topic_dists, doc_lengths, vocab, term_frequency)
    pyLDAvis.display(vis)

    # Stacked bar chart for topic distribution across documents
    topic_proportions = doc_topic_dists / doc_topic_dists.sum(axis=1, keepdims=True)
    df_topics = pd.DataFrame(topic_proportions, columns=[f"Topic {i+1}" for i in range(n_topics)])
    df_topics.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
    plt.title("Topic Distribution Across Documents")
    plt.xlabel("Document")
    plt.ylabel("Proportion")
    plt.legend(title="Topics")
    plt.show()
```

```python
# Separate content for male and female politicians
male_corpus = []
female_corpus = []

male_names = {'donald trump', 'rishi sunak'}
female_names = {'kamala harris', 'elizabeth truss'}

for name, df in factiva_dataframes.items():
    name_lower = name.lower().strip()  # Normalize the key

    # Combine all 'content' rows into the corpus
    combined_texts = df['content_adjectives'].dropna().tolist()  # Get list of all articles

    if name_lower in female_names:
        female_corpus.extend(combined_texts)  # Add all articles
    elif name_lower in male_names:
        male_corpus.extend(combined_texts)  # Add all articles


print(f"Female corpus size: {len(female_corpus)}")
print(f"Male corpus size: {len(male_corpus)}")
```

## **6**.**Limitation & Next-step**

### 6.1 Limitation

**Inherent Media Biases Beyond Gender**

While this study primarily aims to explore gender bias in media coverage, it is important to acknowledge that media bias is influenced by multiple factors beyond gender, including political affiliation, party loyalties, ideological leanings, and personal political stances of the politicians involved. Media outlets often exhibit preferences based on the political alignment of the figures they cover, which can compound the potential for bias. Additionally, the individual political history, public persona, and previous media treatment of each politician may also play a role in shaping coverage, making it challenging to fully isolate gender as the sole factor. In this study, while efforts were made to control for such variables, it remains difficult to entirely exclude these confounding influences.

**Overlapping Coverage and Language Repetition in Major Political Events**

This study focuses on media coverage during major political events, such as presidential elections and government transitions, where the media’s attention is often highly concentrated and specific. During such times, news coverage tends to focus on a limited set of issues, and the language used in reporting can become repetitive, with many articles echoing the same key phrases and narratives. This repetitive nature can limit the ability to detect subtle differences in the portrayal of male and female politicians, as the focus tends to be more on the event itself (e.g., election outcomes, power transitions) rather than on the individual characteristics of the politicians. While this can help create a more comparable context for analysis, it also introduces the challenge of distinguishing nuanced differences in language use and tone that may indicate gender biases.

**Potential Limitations in Data Sources**

The use of Factiva as the primary source for collecting news articles provides access to major, high-quality publications; however, it also carries certain limitations. Factiva’s database is heavily skewed toward mainstream Western media outlets, which may have their own editorial biases. As such, the results of this study may not fully capture the diversity of perspectives that exists within other international or alternative media outlets.

### 6.2 Next Step

In light of the limitations identified, several steps can be taken to refine and expand this study:

**Expanding the Dataset**

Future research could widen the scope of data collection to include additional political figures and time periods, thereby capturing a broader range of media coverage. The inclusion of non-election periods, smaller political events, and media outlets from a broader spectrum of political ideologies would enhance the generalizability of the findings.

**Deepening Analysis of Gender Stereotypes**

While this study examined media coverage through textual analysis, future studies could explore other forms of media representation, such as the use of images, videos, and non-verbal cues in news coverage. A mixed-methods approach that combines qualitative and quantitative analysis of both textual and visual media could offer a more holistic view of how gender bias operates in political discourse.

**Longitudinal Studies**

Conducting a longitudinal analysis that tracks changes in media representation of female politicians over time would provide insights into whether gender bias is diminishing, increasing, or evolving. This could reveal how changes in political culture, societal norms, or media practices influence the portrayal of women in politics.
